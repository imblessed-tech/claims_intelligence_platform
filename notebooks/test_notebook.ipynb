```python
df_f1 = df_f.copy()
```

```python
df_f1["age_group"] = pd.cut(df_f1["age"], 
                                bins = [0, 25, 35, 45, 55, 100],
                                labels = ["youth", "young_adult", "adult", "middle_aged", "senior"]) 

df_f1["bmi_category"] = pd.cut(df_f1["bmi"], 
                                [0, 18.5, 25.0, 30.0, 100],
                                labels = ["underweight", "normal", "overweight", "obese"])
df_f1["smoker_encoded"] = (df_f1["smoker"] == "yes").astype(int)

df_f1["sex_encoded"] = (df_f1["sex"] == "male").astype(int)

age_min = df_f1["age"].min
age_max = df_f1["age"].max
df_f1["age_norm"] = (df_f1 - age_min) /(age_max - age_min)
```
